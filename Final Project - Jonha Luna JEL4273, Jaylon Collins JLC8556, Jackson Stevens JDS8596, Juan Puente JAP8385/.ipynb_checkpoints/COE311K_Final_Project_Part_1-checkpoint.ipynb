# Section 1: Introduction & Model Selection

For this final project, I will utilize Model A, the Spring-Mass System for mechanical oscillations, to model the suspension system of one wheel from a Porsche 911 GT3 R (992). Traditionally, the suspensions of race cars use much stiffer springs, greater damping coefficients (often tuned close to critical damping), and lower masses than a standard passenger car you may drive. This is due to the necessity of the car's chassis to remain stable during aerodynamic loading to maximize performance and keep consistency around the track. Using this as the physical model should give interesting results. First, I must define the variables and parameters that will be used in this project model:

$m$: Sprung mass (the portion of the car's weight supported by this wheel, in kg).

$k$: Spring constant (the stiffness, in N/m).

$c$: Damping coefficient (the shock absorber strength, in N·s/m).

$y(t)$: Vertical displacement of the chassis (in m).

$v(t)$ or $y'(t)$: Vertical velocity (in m/s).

Next, I will define the second-order ODE that will govern the overall behavior of the system. For a suspension system, the car's suspension geometry is modeled as a damped harmonic oscillator, and we use the Second Law ($F=ma$) to form the linear second-order ODE. This model balances the inertial force ($ma$), how the spring's potential energy 'pushes back' when a wheel goes over a bump (this is $ky$), and how the shock absorber resists the overall movement ($c\dot{y}$, $\dot{y}$ being the first time derivative of $y$, or $cv$). This is the full second-order ODE:

$$m\frac{d^{2}y}{dt^{2}}+c\frac{dy}{dt}+ky=0$$

Lastly, we will use Euler's Forward and RK4 methods to solve this system numerically. Because these methods solve first-order equations, we must reduce the second-order ODE to a system of two first-order ODEs. We define state variables for position ($y_1$) and velocity ($y_2$):

Let:$$y_1 = y$$$$y_2 = \frac{dy}{dt}$$

By then taking the time derivative of these state variables, we obtain the system used in our code:

$$\frac{dy_{1}}{dt} = y_{2}$$$$\frac{dy_{2}}{dt} = -\frac{c}{m}y_{2} - \frac{k}{m}y_{1}$$

# Section 3: Numerical Methods Implementation

The two numerical methods that we will be using for analysis are Euler's Forward and Runge-Kutta (RK4). Explicit methods, like Euler’s Forward or RK4, require impractically small time steps to avoid numerical instability. They both use these time steps to approximate the solution by breaking the time domain into discrete steps of size $h$.

For Euler’s Method, it's the simplest numerical procedure for solving ODEs. It uses the slope at the current point $(t_n, y_n)$ to "guess" the value at the next time step:$$y_{n+1} = y_n + h \cdot f(t_n, y_n)$$

In our suspension model, this requires two simultaneous updates: one for position ($y$) and one for velocity ($v$). However, while it is easy to implement, Euler’s method is only first-order accurate, meaning the error is proportional to the step size. It is also conditionally stable, often struggling with oscillatory systems like ours by artificially adding energy to the system if $h$ is too large. For the Runge-Kutta Method, the Fourth-Order Runge-Kutta (RK4) method provides a much higher degree of accuracy by taking a weighted average of four different slopes ($k$ values) within a single time step:

$k_1$: Slope at the start of the interval.

$k_2$: Slope at the midpoint (using $k_1$).

$k_3$: Improved slope at the midpoint (using $k_2$).

$k_4$: Slope at the end of the interval (using $k_3$).

By combining these, RK4 cancels out lower-order error terms, achieving fourth-order accuracy ($O(h^4)$). This makes it significantly more stable and accurate than Euler's method, allowing for larger step sizes while maintaining the physical integrity of the damped oscillation.

```python
import numpy as np
import matplotlib.pyplot as plt

# System Variables/Parameters (all floats to prevent issues with calculation you get with ints)
m = 350.0        # Mass (kg)
k = 150000.0     # Spring constant (N/m)
c = 7000.0         # Damping coefficient (Ns/m)

# Initial conditions
y0 = 0.0      # Initial displacement (m)
v0 = 2.0      # Initial velocity (m/s)
init_state = np.array([y0, v0])

# Time span of the scenario
t0 = 0.0
tf = 1.0       # Simulate for 1 seconds (stiff suspension prevents a lot of oscilation, and race cars are fast over bumps)

def system_ode(t, state): # This is the definition for the entire ODE system that will be used later
# Returns the derivatives for the spring-mass system. state = [y1, y2] = [position, velocity]
    y1, y2 = state
    dy1_dt = y2
    dy2_dt = -(c/m)*y2 - (k/m)*y1
    return np.array([dy1_dt, dy2_dt])

# This is the exact analytical solution for the underdamped system. We're using this to do error analysis with the numerical systems
def exact_solution(t, m, c, k, y0, v0):
    alpha = c / (2*m)
    omega_d = np.sqrt(4*m*k - c**2) / (2*m)
    A = y0
    B = (v0 + alpha * A) / omega_d
    y_exact = np.exp(-alpha * t) * (A * np.cos(omega_d * t) + B * np.sin(omega_d * t))
    return y_exact

# The functions for the numerical solvers
def euler_forward(f, state0, t0, tf, h):
    t_vals = np.arange(t0, tf + h, h)
    states = np.zeros((len(t_vals), len(state0)))
    states[0] = state0
    
    for i in range(1, len(t_vals)):
        states[i] = states[i-1] + h * f(t_vals[i-1], states[i-1])
        
        # This helps protect against issues with overflow and prevents crashing when conducting stability testing when it "blows up" in Section 5.
        if np.any(np.abs(states[i]) > 1e5):
            states[i:] = np.nan
            break      
    return t_vals, states

def rk4(f, state0, t0, tf, h):
    t_vals = np.arange(t0, tf + h, h)
    states = np.zeros((len(t_vals), len(state0)))
    states[0] = state0
    
    for i in range(1, len(t_vals)):
        t = t_vals[i-1]
        y = states[i-1]
        
        k1 = f(t, y)
        k2 = f(t + h/2, y + h*k1/2)
        k3 = f(t + h/2, y + h*k2/2)
        k4 = f(t + h, y + h*k3)
        
        states[i] = y + (h/6) * (k1 + 2*k2 + 2*k3 + k4)  
    return t_vals, states
```

# Section 4: Solutions & Comparison

```python
h_sizes = [.1, .05, .01] # These are the step sizes
colors = ['r', 'b', 'g'] # These are the best colors for visibility and following separate data models. Other colors are not as intuitive and are sometimes hard to see

t_exact = np.linspace(t0, tf, 1000)
y_exact = exact_solution(t_exact, m, c, k, y0, v0)
v_exact = np.gradient(y_exact, t_exact)

# --- FIGURE 1: POSITION ---
# Using a small height (3.5) allows this to tuck into the bottom of a page
plt.figure(figsize=(10, 3.5)) 
plt.plot(t_exact, y_exact, 'k-', lw=4, alpha=0.3, label='Analytical') # The "Shadow" line

for h, color in zip(h_sizes, colors): #Better way to format the loops for the plots according to Gemini
    t_e, s_e = euler_forward(system_ode, init_state, t0, tf, h)
    t_r, s_r = rk4(system_ode, init_state, t0, tf, h)
    plt.plot(t_e, s_e[:, 0], f'{color}--', lw=1.5, label=f'Euler h={h}')
    plt.plot(t_r, s_r[:, 0], f'{color}-', lw=1.5, label=f'RK4 h={h}')

plt.title('Position vs. Time')
plt.ylabel('Displacement (m)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# --- FIGURE 2: VELOCITY ---
plt.figure(figsize=(10, 3.5))
plt.plot(t_exact, v_exact, 'k-', lw=4, alpha=0.3, label='Analytical')

for h, color in zip(h_sizes, colors):
# We solve again inside each subplot section, or store them earlier
    t_e, s_e = euler_forward(system_ode, init_state, t0, tf, h)
    t_r, s_r = rk4(system_ode, init_state, t0, tf, h)
    plt.plot(t_e, s_e[:, 1], f'{color}--', lw=1.5, label=f'Euler h={h}')
    plt.plot(t_r, s_r[:, 1], f'{color}-', lw=1.5, label=f'RK4 h={h}')

plt.title('Velocity vs. Time')
plt.ylabel('Velocity (m/s)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# --- FIGURE 3: PHASE PORTRAIT ---
plt.figure(figsize=(10, 3.5))
plt.plot(y_exact, v_exact, 'k-', lw=4, alpha=0.3, label='Analytical')

for h, color in zip(h_sizes, colors):
    t_e, s_e = euler_forward(system_ode, init_state, t0, tf, h)
    t_r, s_r = rk4(system_ode, init_state, t0, tf, h)
    plt.plot(s_e[:, 0], s_e[:, 1], f'{color}--', lw=1.5, label=f'Euler h={h}')
    plt.plot(s_r[:, 0], s_r[:, 1], f'{color}-', lw=1.5, label=f'RK4 h={h}')

plt.title('Phase Portrait: Velocity vs. Position')
plt.xlabel('Displacement (m)')
plt.ylabel('Velocity (m/s)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

We tested both Euler’s Forward and RK4 methods across three step sizes ($h=.1, .05, .01$) to simulate the response of the Porsche after a 2 $m/s$ impact with its front suspension. At every tested step size, the RK4 solution remains almost perfectly aligned with the analytical curve. This shows its strenght with systems wih oscillations due to its fourth-order accuracy. Even with a large $h=.1$, it captures the stiff suspension's motion without visible error. However, Euler’s method is highly sensitive to the step size $h$, which is a known limitation of the solution. At $h=.01$, it is mostly stable and accurate, but at $h=.1$, the method fails completely. The $h=.1$ Euler solution exhibits extreme numerical instability. In the Position vs. Time plot, the car reaches a non-physical displacement of over 20 $\text{ meters}$, which, physically, is an impossible distance for a car's suspension. A similar disparity is shown in the velocity vs. time graph, which shows unrealistic speeds of over 200 $m/s$ and nearly -800 $m/s$. In a real-world car's suspension, energy dissipates, causing the displacement to decay toward zero. However, when the Euler step size is too large ($h > h_{crit}$), the method adds more "energy" to the system at each step faster than the physical damper can remove. As a result, instead of a steady decay, the unstable solution shows the car "bouncing" higher with every oscillation, meaning the car is basically gaining kinetic energy from nowhere, eventually leading to the "blow-up" where displacement exceeds the physical limitations of the vehicle. Furthermore, in the Phase Portrait, while the stable solutions spiral inward toward the equilibrium point (0,0), the unstable Euler solution spirals outward. This indicates that the numerical error is artificially increasing the system's total energy, causing the "car" to bounce higher and faster with every step rather than decaying. This comparison suggests that for a oscilitory system as stiff as the Porsche ($k = 150,000\text{ N/m}$), a step size of $h=.1$ is too "strong" for Euler's method to resolve the high-frequency oscillation, leading to divergence.

# Section 5: Stability Analysis

```python
# Instability Demonstration
h_unstable = .025 # Found the insability to be .02, and I exaggerated the step size value to show where Euler starts to "blow up". Found using trial and error

t_unstable, states_unstable = euler_forward(system_ode, init_state, t0, tf, h_unstable)

plt.figure(figsize=(6, 2))
plt.plot(t_exact, y_exact, 'k-', label='Analytical')
plt.plot(t_unstable, states_unstable[:, 0], 'r--', label=f'Euler (Unstable) h={h_unstable}')
plt.title('Euler Method Instability')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.ylim(-.2, .2)
plt.legend()
plt.grid(True)
plt.show()

# This is for error quantification (the Log-Log Plot)
h_test_vals = np.logspace(-4, -2, 10) # <--- Might need to adjust these bounds based on your stability limit
euler_error = []
rk4_error = []

for h in h_test_vals:
    t_e, s_e = euler_forward(system_ode, init_state, t0, tf, h)
    y_exact_e = exact_solution(t_e, m, c, k, y0, v0)
    err_e = np.nanmean(np.abs(s_e[:, 0] - y_exact_e))
    euler_error.append(err_e)
    
    t_r, s_r = rk4(system_ode, init_state, t0, tf, h)
    y_exact_r = exact_solution(t_r, m, c, k, y0, v0)
    err_r = np.nanmean(np.abs(s_r[:, 0] - y_exact_r))
    rk4_error.append(err_r)

plt.figure(figsize=(6, 4))
plt.loglog(h_test_vals, euler_error, 'ro-', label='Euler Error')
plt.loglog(h_test_vals, rk4_error, 'bs-', label='RK4 Error')

h_ref = np.array([1e-4, 1e-2])
plt.loglog(h_ref, 1e1 * h_ref**1, 'k--', label='Slope = 1 (O(h))')
plt.loglog(h_ref, 1e1 * h_ref**4, 'g--', label='Slope = 4 (O(h^4))')

plt.title('Error Convergence: Log-Log Plot')
plt.xlabel('Step Size (h)')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.grid(True)
plt.show()
```

For the Porsche model, we found a stark contrast in stability between Euler’s Forward and RK4. Firstly, Euler’s method is only conditionally stable. For our high-stiffness system ($k = 150,000$ N/m), the maximum stable step size ($h_{crit}$) was approximately .01, so as evidenced in the first plot, even at just $h = 0.025$, the Euler solution (the red dashed line) begins to overshoot the analytical peak and oscillates with increasing error. RK4, on the other hand, is significantly more robust as it remains stable and accurate at $h=0.1$, confirming that its stability region is much larger than Euler’s for this specific system. The Log-Log plot also verifies the mathematical integrity of our implementation. The red Euler’s data points have the same slope as the black dashed reference line (Slope = 1), confirming that Euler’s Forward is a first-order method ($O(h)$); if you cut the step size in half, the error generally cuts in half. Meanwhile, the blue RK4 points align with the green dashed reference line (Slope = 4), confirming that it is a fourth-order method ($O(h^4)$). Furthermore, at a step size of $h = 10^{-3}$, the RK4 error is approximately $10^{-11}$, while the Euler error is $10^{-4}$, meaning RK4 provides nearly ten million times more precision at the same step size.

# Section 6: Conclusions

Ultimately, for running a simulation of the Porsche 911 GT3 R suspension, the Fourth-Order Runge-Kutta (RK4) method performs significantly better than Euler’s Forward method. While both methods are capable of approximating the damped harmonic oscillation at very fine step sizes, RK4 maintained near-perfect alignment with the analytical solution across all tested parameters. Conversely, Euler’s method proved to be highly fragile, failing at multiple points, such as requiring a step size at least five times smaller than RK4 just to maintain basic numerical stability. Using a method like RK4 does have its tradeoffs, however. Euler’s Method does benefit from having a lower computational cost per individual step (requiring only one function evaluation), while RK4 is more expensive per step (requiring four function evaluations). However, when precision and real-life correlation is the goal like in motorsport, the efficiency and stability of RK4 makes it the obvious choice despite its complexity. For high-stiffness mechanical systems like a GT3-class race car, we recommend to avoid simple explicit solvers like Euler’s Forward for "stiff" ODEs, as the step size requirements become impractically small and to instead use adaptive step-size solvers like RK4 engineering tools to ensure that numerical "blow-ups" do not provide false data, and to run convergence/stability/correlation studies before trusting a simulation’s results for car setup.
