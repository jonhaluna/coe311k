**Section 1: Introduction**

**Description of the System**

We chose Option A, an RC circuit with a fast transient with a voltage dependent on time. The system models how the voltage across the capacitor changes when a varying input is introduced. This is a stiff system because there are two different time scales.

**Governing ODE**

The governing equation is: 

$\frac{dV}{dt} = -\frac{1}{RC} V + \frac{1}{RC} V_{\text{source}}(t)$ 

With parameters:

R = $100$ Ω

C = $10^{-9}$ F

$V_{\text{source}}(t) = 5sin(t)$

the equation simplifies to:

$\frac{dV}{dt} = -10^{7} V + 5 \cdot 10^{7} \sin(t)$

**Physical Meaning**

V(t): Voltage across the capacitor

R: Resistance, how easily current can flow

C: Capacitance, how well the capacitor can store charge

RC: Time constant of the circuit, how quickly the system responds

$V_{\text{source}}(t)$\: External driving voltage

The time constant is: $RC = 10^{-7}$ seconds, which is a very fast response

**Why this system is important in practice**

RC circuits are crucial in many different applications in engineering like signal fitering and timing circuits. They are especially important in high-speed electronics, where a fast transient behavior happens a lot. Studying this system is important because stiff equations come up a lot in real-world situations. Using efficient numerical methods let us have accurate and stable simulation without needing really small time step sizes.

**Section 2: Demonstration of Stiffness**

**Euler's Forward Attempt**

Using Euler's Forward method:

$V_{n+1} = V_n + h\left(-10^{7} V_n + 5 \cdot 10^{7} \sin(t_n)\right)$

and a step size of $h = 0.1$, we see that the solution becomes unstable and diverges instead of following a sinusoidal behavior.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, V):
    return -1e7 * V + 5e7 * np.sin(t)

def euler_forward(f, t0, V0, h, t_end):
    N = int((t_end - t0) / h)
    t = np.linspace(t0, t_end, N+1)
    V = np.zeros(N+1)
    
    V[0] = V0
    
    for i in range(N):
        V[i+1] = V[i] + h * f(t[i], V[i])
    
    return t, V

t0 = 0
V0 = 0
h = 0.1
t_end = 2

t_vals, V_vals = euler_forward(f, t0, V0, h, t_end)

plt.figure()
plt.plot(t_vals, V_vals)
plt.title("Euler Forward Method (h = 0.1)")
plt.xlabel("Time t")
plt.ylabel("Voltage V(t)")
plt.grid()
plt.show()
```

**Stability Analysis**

For a linear system, Euler's forward stability requires that 

$\left|1 + h\lambda \right| < 1$

For this system, $\lambda = -10^{7}$

So, $h < 2 \cdot 10^{-7}$

**Stiffness Ratio**

This non-linear system has two different time values, 
$t_{\text{fast}} = 10^{-7}$ and $t_{\text{slow}}$ = $\frac{2\pi}{\omega}$, 
where $\omega = 1$, so $t_{\text{slow}}$ = ${2\pi}$
The stiffness ratio is $\frac{t_{\text{slow}}}{t_{\text{fast}}} = \frac{2\pi}{10^{-7}} = 2\pi \cdot 10^{7}$ which tells us that the system is very stiff. The stiffness is so high because the system has a decaying transient and a slowly varying function.

**Why Implicit Methods are Necessary**

If you tried using Forward Euler to solve this, you would find that the number of steps required is $5 \cdot 10^7$, which is computationally expensive and inefficient for a system like this. So, implicit methods like Euler's Backward work much better for this system because they allow for large step sizes and will reduce computational costs.

**Section 3: Mathematical Setup**

**Residual Function**

Euler's Backward method: $V_{n+1} = V_n + h\, f(t_{n+1}, V_{n+1})$

Residual function: $g(V_{n+1}) = V_{n+1} - V_n - h\,f(t_{n+1}, V_{n+1})$

For the ODE, we can set $f(t, V) = -10^{7} V + 5 \cdot 10^{7} \sin(t)$,
and by substituting and simplifying we get: 
$g(V_{n+1}) = (1 + 10^{7}h)\,V_{n+1} - V_n - 5 \cdot 10^{7} h \sin(t_{n+1})$

**Analytical Derivation**

Using the formula: $g'(y_{n+1}) = 1 - h \frac{\partial f}{\partial y}$ where $\frac{\partial f}{\partial V} = -10^{7}$ we find that
$g'(V_{n+1}) = 1 - h(-10^{7}) = 1 + 10^{7}h$

**Newton-Raphson Solver**

To solve $g(y_{n+1}) = 0$, we use the Newton-Raphson iteration:
$V_{n+1}^{(k+1)} = V_{n+1}^{(k)} - \frac{g\!\left(V_{n+1}^{(k)}\right)}{g'\!\left(V_{n+1}^{(k)}\right)}$
Substituting expressions, we get 
$V_{n+1}^{(k+1)} = V_{n+1}^{(k)} - \frac{(1 + 10^{7}h)\,V_{n+1}^{(k)} - V_n - 5 \cdot 10^{7}h \sin(t_{n+1})}{1 + 10^{7}h}$

Because this is a linear system, Newton's method converges after one iteration

**Section 4: Implementation**

**Step Size Comparison**

From Demonstration of Stiffness, the maximum stable step size for Euler's Forward is $ 2 \cdot 10^{-7}$
Euler's Backward is unconditionally stable, which means that it stays stable for any step size

$\begin{array}{|c|c|c|}\hline
\text{Method} & \text{Lower Bound} & \text{Upper Bound} \\ \hline
\text{Forward Euler} & 0 & 2\times10^{-7} \\ \hline
\text{Backward Euler} & 0 & \infty \\ \hline
\end{array}$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, V):
    return -1e7 * V + 5e7 * np.sin(t)

def euler_forward(t0, V0, h, t_end, max_val=1e6):
    N = int((t_end - t0) / h)
    t = np.linspace(t0, t_end, N+1)
    V = np.zeros(N+1)
    V[0] = V0

    for i in range(N):
        V[i+1] = V[i] + h * f(t[i], V[i])

        # stops if unstable
        if abs(V[i+1]) > max_val:
            return t[:i+2], V[:i+2]

    return t, V

# Unstable

t1, V1 = euler_forward(0, 0, 0.1, 0.3)

# Stable

t2, V2 = euler_forward(0, 0, 1e-5, 5e-5)

#Plot

plt.figure(figsize=(6,8))

plt.subplot(2,1,1)
plt.plot(t1, V1)
plt.title("Forward Euler (h = 0.1, Unstable)")
plt.xlabel("Time (s)")
plt.ylabel("V(t)")
plt.grid()

plt.subplot(2,1,2)
plt.plot(t2, V2)
plt.title("Forward Euler (h = 1e-5, Delayed Instability")
plt.xlabel("Time (s)")
plt.ylabel("V(t)")
plt.grid()

plt.tight_layout()
plt.show()
```

For a smaller step size, the solution appears stable, but quickly becomes unstable. This highlights the step size restrictions of Euler's Forward method with stiff systems.

**Accuracy vs. Computational Cost**

```python
import numpy as np

# ODE
def f(t, V):
    return -1e7 * V + 5e7 * np.sin(t)

# Derivative of residual
def g_prime(h):
    return 1 + 1e7 * h

# Newton-Raphson solver
def newton_solver(Vn, tn1, h, tol=1e-8, max_iter=20):
    V = Vn
    gp = g_prime(h)

    iterations = 0
    func_evals = 0

    for _ in range(max_iter):
        g = V - Vn - h * f(tn1, V)
        func_evals += 1

        V_new = V - g / gp
        iterations += 1

        if abs(V_new - V) < tol:
            return V_new, iterations, func_evals

        V = V_new

    raise RuntimeError("Newton did not converge")

# Backward Euler cost
def backward_euler_cost(h, t_end=1):
    N = int(t_end / h)

    V = 0
    total_iters = 0
    total_fevals = 0

    for i in range(N):
        tn1 = (i + 1) * h
        V, iters, fevals = newton_solver(V, tn1, h)

        total_iters += iters
        total_fevals += fevals

    avg_iters = total_iters / N

    return N, avg_iters, total_fevals

h_values = [1e-2, 1e-3, 1e-4]

print("Backward Euler Cost:\n")

for h in h_values:
    N, avg_iters, total_fevals = backward_euler_cost(h)

    print(f"Step size h = {h}")
    print(f"  Total time steps: {N}")
    print(f"  Average Newton iterations per step: {avg_iters:.2f}")
    print(f"  Total function evaluations: {total_fevals}\n")

# Comparison of Forward Euler
h_forward = 2e-7  # stability limit
N_forward = int(1 / h_forward)

print("Forward Euler Cost:\n")
print(f"Step size h = {h_forward}")
print(f"  Total time steps: {N_forward}")
print(f"  Total function evaluations: {N_forward}")
print("  (No Newton iterations required)\n")
```

This shows that using Euler's Backward method uses much less steps while staying stable.

**Convergence Behavior**

**Newton Iterations vs Time**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, V):
    return -1e7 * V + 5e7 * np.sin(t)

def g_prime(h):
    return 1 + 1e7 * h

def newton_solver(Vn, tn1, h, tol=1e-8, max_iter=20):
    V = Vn
    gp = g_prime(h)
    iterations = 0

    for _ in range(max_iter):
        g = V - Vn - h * f(tn1, V)
        V_new = V - g / gp
        iterations += 1

        if abs(V_new - V) < tol:
            return V_new, iterations

        V = V_new

    return V, iterations

def run_sim(h, t_end=1):
    N = int(t_end / h)
    t = np.linspace(0, t_end, N+1)
    V = np.zeros(N+1)
    iters = np.zeros(N)

    for i in range(N):
        V[i+1], k = newton_solver(V[i], t[i+1], h)
        iters[i] = k

    return t, iters

# Simulation
t, iters = run_sim(1e-3)

print("Newton Iterations vs Time (h=1e-3):")
print(f"  Average iterations: {np.mean(iters):.2f}")
print(f"  Max iterations: {np.max(iters)}\n")

plt.figure()
plt.plot(t[:-1], iters)
plt.title("Newton-Raphson Iterations vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Iterations per Step")
plt.grid()
plt.show()
```

The number of iterations is constant, showing that the convergence is fast and stable.

**Dependence on Step Size**

```python
h_values = [1e-2, 1e-3, 1e-4]

plt.figure()
print("Step Size Dependence:")

for h in h_values:
    t, iters = run_sim(h)
    print(f"  h = {h}: avg = {np.mean(iters):.2f}, max = {np.max(iters)}")
    plt.plot(t[:-1], iters, label=f"h={h}")

plt.legend()
plt.title("Newton Iterations vs Time (Different Step Sizes)")
plt.xlabel("Time (s)")
plt.ylabel("Iterations")
plt.grid()
plt.show()
```

From this, we see the number of iterations is idependent of step size, because the system is linear.

**Dependence on Initial Guess**

```python
def newton_with_guess(Vn, tn1, h, guess_type="good"):
    if guess_type == "good":
        V = Vn
    else:
        V = 0  # bad guess

    gp = g_prime(h)
    iterations = 0

    for _ in range(20):
        g = V - Vn - h * f(tn1, V)
        V_new = V - g / gp
        iterations += 1

        if abs(V_new - V) < 1e-8:
            return iterations

        V = V_new

    return iterations

def compare_guesses(h=1e-3, t_end=1):
    N = int(t_end / h)
    t = np.linspace(0, t_end, N+1)

    it_good = np.zeros(N)
    it_bad = np.zeros(N)
    V = np.zeros(N+1)

    for i in range(N):
        it_good[i] = newton_with_guess(V[i], t[i+1], h, "good")
        it_bad[i] = newton_with_guess(V[i], t[i+1], h, "bad")
        V[i+1], _ = newton_solver(V[i], t[i+1], h)

    return t, it_good, it_bad

t, it_good, it_bad = compare_guesses()

print("Initial Guess Comparison:")
print(f"  Good guess avg iterations: {np.mean(it_good):.2f}")
print(f"  Bad guess avg iterations: {np.mean(it_bad):.2f}\n")

plt.figure()
plt.plot(t[:-1], it_good, label="Good Guess (Vn)")
plt.plot(t[:-1], it_bad, label="Bad Guess (0)")
plt.legend()
plt.title("Effect of Initial Guess on Convergence")
plt.xlabel("Time (s)")
plt.ylabel("Iterations")
plt.grid()
plt.show()
```

This shows us that convergence stays fast and stable.

**Convergence Dependence on System Nonlinearity**

The system is linear, so Newton-Raphson converges after one iteration.

```python
def f_nonlinear(t, V):
    return -1e7 * V**2 + 5e7 * np.sin(t)

def newton_nonlinear(Vn, tn1, h):
    V = Vn
    iterations = 0

    for _ in range(20):
        g = V - Vn - h * f_nonlinear(tn1, V)
        gp = 1 - h * (-2e7 * V) # derivative
        V_new = V - g / gp
        iterations += 1

        if abs(V_new - V) < 1e-8:
            return iterations

        V = V_new

    return iterations

def nonlinear_test(h=1e-3, t_end=1):
    N = int(t_end / h)
    t = np.linspace(0, t_end, N+1)
    V = np.zeros(N+1)
    iters = np.zeros(N)

    for i in range(N):
        iters[i] = newton_nonlinear(V[i], t[i+1], h)
        V[i+1] = V[i]  # simplified

    return t, iters

t, iters_nl = nonlinear_test()

print("Nonlinear System:")
print(f"  Average iterations: {np.mean(iters_nl):.2f}")
print(f"  Max iterations: {np.max(iters_nl)}\n")

plt.figure()
plt.plot(t[:-1], iters_nl)
plt.title("Newton Iterations (Nonlinear System)")
plt.xlabel("Time (s)")
plt.ylabel("Iterations")
plt.grid()
plt.show()
```

This tells us that for nonlinear systems, the number of iterations increases and is dependent on step size and initial guess.

**Solution Quality**

**Errors of Reference Solution**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, V):
    return -1e7 * V + 5e7 * np.sin(t)

def g_prime(h):
    return 1 + 1e7 * h

def newton_solver(Vn, tn1, h, tol=1e-10, max_iter=20):
    V = Vn
    gp = g_prime(h)

    for _ in range(max_iter):
        g = V - Vn - h * f(tn1, V)
        V_new = V - g / gp

        if abs(V_new - V) < tol:
            return V_new

        V = V_new

    raise RuntimeError("Newton did not converge")

def backward_euler(h, t_end=1):
    N = int(t_end / h)
    t = np.linspace(0, t_end, N+1)
    V = np.zeros(N+1)

    for i in range(N):
        V[i+1] = newton_solver(V[i], t[i+1], h)

    return t, V

h_ref = 1e-5
t_ref, V_ref = backward_euler(h_ref)

h_vals = [1e-2, 1e-3, 1e-4]
errors = []

for h in h_vals:
    t, V = backward_euler(h)

    V_ref_interp = np.interp(t, t_ref, V_ref)

    mask = t > 0.01
    error = np.max(np.abs(V[mask] - V_ref_interp[mask]))
    errors.append(error)

print("Error vs Step Size):")
for h, e in zip(h_vals, errors):
    print(f"  h = {h}, max error = {e:.3e}")

# Plot
plt.figure()
plt.loglog(h_vals, errors, marker='o')
plt.title("Error vs Step Size (Backward Euler)")
plt.xlabel("Step Size h")
plt.ylabel("Maximum Error")
plt.grid()
plt.show()
```

**Section 6: Performance Analysis**

**Computational Cost Analysis**

```python
def forward_euler_cost(h, t_end):
    N = int(t_end / h)
    return {
        "h": h,
        "steps": N,
        "function_evals": N
    }

def backward_euler_cost(h, t_end):
    t, V, iters = backward_euler_with_iters(h, t_end)
    N = len(iters)

    return {
        "h": h,
        "steps": N,
        "avg_NR_iters": np.mean(iters),
        "total_NR_iters": np.sum(iters),
        "function_evals": np.sum(iters)
    }

fe_cost = forward_euler_cost(1e-7, 1e-3)
be_cost = backward_euler_cost(1e-2, 1e-3)

print("Forward Euler Cost:", fe_cost)
print("Backward Euler Cost:", be_cost)
```

**Accuracy-Efficiency Takeoff**

```python
h_values = [1e-2, 1e-3, 1e-4]

steps = []
avg_iters = []

for h in h_values:
    t, V, iters = backward_euler_with_iters(h, t_end=1)
    steps.append(len(t)-1)
    avg_iters.append(np.mean(iters))

plt.figure()
plt.plot(h_values, steps, marker='o')
plt.xscale('log')
plt.title("Number of Time Steps vs Step Size (Backward Euler)")
plt.xlabel("Step Size h")
plt.ylabel("Number of Steps")
plt.grid()
plt.show()
```

As step size decreases, the number of required steps increases significantly, which raises the computational cost. However, Newton-Raphson iterations are a constant one iteration per step, so the cost per step is very small. This makes Euler's Backward efficient even when small steps are used for high accuracy solutions. 

**Practical Recommendations**

Don't use Euler's Forward for stiff systems because of its step size requirements. We recommend using Euler's Backward because it stays stable unconditionally, reduces the number of required steps, and for this linear system, Newton-Raphson converges in one iteration.

**Section 7: Conclusions**

For this project, we evaluated the numerical solutions of a stiff ordinary differential equation using both explicit and implicit methods. The results show that Euler's Forward method is impractical for stiff systems because of its step size restrictions. Stability analysis showed that Euler's Forward needs really small step sizes to not diverge, making it very computationally expensive and inefficient. Euler's Backward however, remained stable for all step sizes and produced solutions with accuracy even with large h values. 
The Newton-Raphson method used in Euler's Backward converged in a single iteration per step because the system was linear. This gave us a very low computational cost per step, making the implicit method stable and efficient. This combination makes Euler's Backward very good for use in stiff problems. Euler's Forward and other explicit methods are good for non stiff systems where the solution changes gradually and constraints aren't as restrictive. But for stiff systems with fast transients and varying time scales, like the one we chose, implicit methods like Euler's Backward are much better. 
One of the main challenges encountered in this project was managing the instability in Euler's Forward simulations. Large step sizes caused it to diverge, while small step sizes had a very large computational cost. Another challenge was making sure the Newton-Raphson was consistent across different sections of the code. 
Overall, we learned that implicit methods are an efficient approach for solving stiff differential equations, while explicit methods are limited due to their stability constraints.
