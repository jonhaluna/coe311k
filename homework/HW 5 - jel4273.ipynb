```python
import numpy as np
import matplotlib.pyplot as plt

def ls(X,Y):
    n=len(X)
    m=(n*np.sum(X*Y)-np.sum(X)*np.sum(Y))/(n*np.sum(X*X)-(np.sum(X))**2)
    b=(np.sum(Y)-m*np.sum(X))/n
    return m,b

X=np.sort(np.random.rand(20))
Y=np.random.rand(20)

m,b=ls(X,Y)
A=np.vstack([X,np.ones(len(X))]).T
m2,b2=np.linalg.lstsq(A,Y,rcond=None)[0]

print("My:",m,b)
print("Numpy:",m2,b2)

x=np.arange(0,1.01,0.01)
mt=np.random.uniform(1,5)
bt=np.random.uniform(0,3)
y=mt*x+bt+np.random.uniform(-1,1,len(x))

m,b=ls(x,y)
A=np.vstack([x,np.ones(len(x))]).T
m2,b2=np.linalg.lstsq(A,y,rcond=None)[0]

yp=m*x+b
r=y-yp

print("Original:",mt,bt)
print("My:",m,b)
print("Numpy:",m2,b2)

print("Largest residual:",r[np.argmax(abs(r))])
print("Smallest residual:",r[np.argmin(abs(r))])

plt.scatter(x,y,s=10)
plt.plot(x,yp)
plt.show()
```
