```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random

def approximatepi(samples):
    counter = 0
    for i in range(0,samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            counter = counter + 1
    return (4*counter)/samples

sampleslist = []
errorlist = []

for samples in range(10,1000,20):
    pi_estimate = approximatepi(samples)
    error = abs(pi_estimate - np.pi) / np.pi
    sampleslist.append(samples)
    errorlist.append(error)    
    
plt.plot(sampleslist, errorlist)
plt.xlabel("Number of Samples")
plt.ylabel("Error Value")
plt.title("Monte Carle pi Approximation Error")
plt.show()
```
