```python
import numpy as np
import matplotlib.pyplot as plt
import math

def taylor_sin(x, n):
    return sum(((-1)**k)*x**(2*k+1)/math.factorial(2*k+1) for k in range(n))

x = np.arange(0, 10.1, 0.1)
exact = np.sin(x)

for n in range(1,5):
    approx = taylor_sin(x, n)
    error = abs(approx - exact)
    plt.plot(x, error, label=f"{n} term")

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Error for sin(x)")
plt.legend()
plt.grid()
plt.show()
```

The truncation error is small at x=0, and it increases as x becomes larger and decreases as x becomes smaller. The error does not decrease uniformly when more terms are added. In fact, when more terms are added for larger values of x, the error can increase far more quickly.
