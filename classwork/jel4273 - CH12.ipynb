```python
import numpy as np
import time
import matplotlib.pyplot as plt

def naive_multiply(A, B):
    n = len(A)
    result = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

sizes = [3]
for _ in range(7):
    sizes.append(sizes[-1]*2)

naive_times = []
numpy_times = []

for size in sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start = time.time()
    naive_multiply(A.tolist(), B.tolist())
    naive_times.append(time.time() - start)

    start = time.time()
    np.dot(A, B)
    numpy_times.append(time.time() - start)

plt.plot(range(1, 9), naive_times, label='Naive')
plt.plot(range(1, 9), numpy_times, label='NumPy')
plt.xlabel('Matrix Number')
plt.ylabel('Runtime')
plt.legend()
plt.show()
```
