```python
import numpy as np
import matplotlib.pyplot as plt

x_start, x_end = 0, 2

def f(x, y):
    return -2 * y

first_run = None

plt.figure(figsize=(10,6))

for h in np.arange(0.01, 1.05, 0.05):
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1

    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    if first_run is None:
        first_run = y_values

    # Compute error vs first run
    min_len = min(len(first_run), len(y_values))
    pointwise_error = np.abs(y_values[:min_len] - first_run[:min_len])
    total_error = np.sum(pointwise_error)

    print(f"Step size h = {h:.2f}, Pointwise error = {pointwise_error}, Total accumulated error = {total_error:.5f}")

    plt.plot(x_values, y_values, label=f"h={h:.2f}")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method Approximation for dy/dx = -2y")
plt.legend()
plt.show()
```
