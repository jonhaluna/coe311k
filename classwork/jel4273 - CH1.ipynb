```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

def f1(x,y):
    return (x*x+y*y)
x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x),'r')

plt.subplot(2,2,2)
y_line=np.cos(x)
plt.scatter(x,y_line,10,'g')

plt.subplot(2,2,3)
plt.hist(x,100)

fig=plt.figure()
ax = fig.add_subplot(2, 2, 4, projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y_line')
ax.set_zlabel('z');
ax.view_init(45, 0)

plt.show()
```
