```python
import numpy as np
import time

def jacobi(A,b,x0,tol,maxit):
    n=len(b)
    x=x0.copy()
    start=time.time()
    for k in range(maxit):
        x_new=x.copy()
        for i in range(n):
            s=sum(A[i][j]*x[j] for j in range(n) if j!=i)
            x_new[i]=(b[i]-s)/A[i][i]
        if np.linalg.norm(x_new-x,ord=np.inf)<tol:
            return x_new,k+1,time.time()-start,True
        x=x_new
    return x,maxit,time.time()-start,False

A=np.array([[0.582745,0.48,0.10,0.0,0.0],
            [0.48,1.044129,0.46,0.10,0.0],
            [0.10,0.46,1.10431,0.44,0.10],
            [0.0,0.10,0.44,0.963889,0.42],
            [0.0,0.0,0.10,0.42,0.522565]],dtype=float)

b=np.array([1.162745,2.084129,2.20431,1.923889,1.042565],dtype=float)
x0=np.zeros_like(b)

sol,it,t,conv=jacobi(A,b,x0,1e-6,10000)

print("Solution:",sol)
print("Iterations:",it)
print("Time:",t,"seconds")
print("Converged:",conv)
```

```python
import numpy as np
import time

def gauss_seidel(A,b,x0,tol,maxit):
    n=len(b)
    x=x0.copy()
    start=time.time()
    for k in range(maxit):
        x_new=x.copy()
        for i in range(n):
            s1=sum(A[i][j]*x_new[j] for j in range(i))
            s2=sum(A[i][j]*x[j] for j in range(i+1,n))
            x_new[i]=(b[i]-s1-s2)/A[i][i]
        if np.linalg.norm(x_new-x,ord=np.inf)<tol:
            return x_new,k+1,time.time()-start,True
        x=x_new
    return x,maxit,time.time()-start,False

A=np.array([[0.582745,0.48,0.10,0.0,0.0],
            [0.48,1.044129,0.46,0.10,0.0],
            [0.10,0.46,1.10431,0.44,0.10],
            [0.0,0.10,0.44,0.963889,0.42],
            [0.0,0.0,0.10,0.42,0.522565]],dtype=float)

b=np.array([1.162745,2.084129,2.20431,1.923889,1.042565],dtype=float)
x0=np.zeros_like(b)

sol,it,t,conv=gauss_seidel(A,b,x0,1e-6,10000)

print("Solution:",sol)
print("Iterations:",it)
print("Time:",t,"seconds")
print("Converged:",conv)
```
