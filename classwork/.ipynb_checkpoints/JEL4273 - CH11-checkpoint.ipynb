```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([0,1,2,3,4,5])
y = np.array([0,1,0,1,0,1])

cs = CubicSpline(x,y,bc_type='natural')

p = np.polyfit(x,y,3)
poly = np.poly1d(p)

x2 = np.linspace(0,5,100)

y_spline = cs(x2)
y_poly = poly(x2)

ys = cs(x)
yp = poly(x)

mse_spline = np.mean((y-ys)**2)
mse_poly = np.mean((y-yp)**2)

ss = np.sum((y-np.mean(y))**2)

r2_spline = 1 - np.sum((y-ys)**2)/ss
r2_poly = 1 - np.sum((y-yp)**2)/ss

print(mse_spline)
print(r2_spline)
print(mse_poly)
print(r2_poly)

if mse_spline < mse_poly:
    print("CubicSpline is better")
else:
    print("Polyfit is better")

plt.scatter(x,y)
plt.plot(x2,y_spline)
plt.plot(x2,y_poly)
plt.show()
```
