```python
import numpy as np
import matplotlib.pyplot as plt

f = np.sin
a, b = 0, 2*np.pi
N = 100
x = np.linspace(a, b, N)
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
forward_diff = (f(x + h) - f(x)) / h

# Backward Difference Approximation
backward_diff = (f(x) - f(x - h)) / h

# Plot the approximation and the true derivative
plt.figure(figsize=(8,6))
plt.plot(x, np.cos(x), 'k', label='cos(x)')
plt.plot(x, forward_diff, 'r', label='Forward Difference')
plt.plot(x, backward_diff, 'b', label='Backward Difference')
plt.plot(x, df_dx, 'g', label='Central Difference')
plt.legend()
plt.show()
```
