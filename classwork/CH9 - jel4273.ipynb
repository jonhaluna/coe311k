```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5])
Y = np.array([1.2,2.8,3.6,4.5,5.1])

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print("Slope:", m)
print("Intercept:", c)

plt.scatter(X, Y)

x_line = np.linspace(min(X), max(X), 100)
y_line = m*x_line + c

plt.plot(x_line, y_line)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5])
Y = np.array([1.2,2.8,3.6,4.5,5.1])

np.random.shuffle(X)
np.random.shuffle(Y)

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print("Slope:", m)
print("Intercept:", c)

plt.scatter(X, Y)

x_line = np.linspace(min(X), max(X), 100)
y_line = m*x_line + c

plt.plot(x_line, y_line)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,1.01,0.01)

m_true = 2
c_true = 1

noise = np.random.uniform(-0.1,0.1,len(x))
y = m_true*x + c_true + noise

A = np.vstack([x, np.ones(len(x))]).T
m,c = np.linalg.lstsq(A,y,rcond=None)[0]

print("True slope:", m_true)
print("True intercept:", c_true)

print("Estimated slope:", m)
print("Estimated intercept:", c)

plt.scatter(x,y)

x_line = np.linspace(0,1,100)
y_line = m*x_line + c

plt.plot(x_line,y_line)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
