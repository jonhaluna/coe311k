```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,1.01,0.01)
y=3*x+2+np.random.uniform(-1,1,len(x))

A=np.vstack([x,np.ones(len(x))]).T
m,b=np.linalg.lstsq(A,y,rcond=None)[0]
yl=m*x+b

p=np.polyfit(x,y,3)
yp=np.polyval(p,x)

mse_l=np.mean((y-yl)**2)
mse_p=np.mean((y-yp)**2)

r2_l=1-np.sum((y-yl)**2)/np.sum((y-np.mean(y))**2)
r2_p=1-np.sum((y-yp)**2)/np.sum((y-np.mean(y))**2)

print("LS",mse_l,r2_l)
print("Poly",mse_p,r2_p)

if mse_l<mse_p: print("LS better")
else: print("Poly better")

plt.scatter(x,y,s=10)
plt.plot(x,yl)
plt.plot(x,yp)
plt.show()
```

```python
import numpy as np

x=np.arange(0,1.01,0.01)
y=3*x+2+np.random.uniform(-1,1,len(x))

for d in range(1,10):
    p=np.polyfit(x,y,d)
    yp=np.polyval(p,x)

    mse=np.mean((y-yp)**2)
    r2=1-np.sum((y-yp)**2)/np.sum((y-np.mean(y))**2)

    print(d,mse,r2)
```

```python
import numpy as np

x=np.arange(0,1.01,0.01)
y=3*x+2+np.random.uniform(-1,1,len(x))

V=np.vander(x,4)
c=np.linalg.lstsq(V,y,rcond=None)[0]
yp=V@c

mse=np.mean((y-yp)**2)
r2=1-np.sum((y-yp)**2)/np.sum((y-np.mean(y))**2)

print("Vandermonde",mse,r2)
```
