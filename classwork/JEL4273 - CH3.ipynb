```python
import math
import random
import numpy as np

def approximation(samples):
    counter = 0
    for i in range(0,samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            counter = counter + 1
    return (4*counter)/samples
```

```python
for i in range(0,100):
    calc_pi = approximation(1000)
    true_pi = np.pi
    true_error = abs(true_pi - calc_pi)
    relative_error = abs(true_pi - calc_pi/(true_pi))
    
    print(true_error, relative_error)
```
