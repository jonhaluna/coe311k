```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[-2, 1],
              [0, -3]])
y0 = np.array([2, 1])
t_final = 5
hs = [0.1, 0.4, 0.8, 1.0]

for h in hs:
    M = np.eye(2) + h*A
    eig = np.linalg.eigvals(M)
    stable = np.all(np.abs(eig) < 1)

    print(f"\nStep size: {h}")
    print("Eigenvalues:", eig)
    print("Solution is Stable" if stable else "Solution may not be Entirely Stable ;)")

    n = int(t_final/h)
    y = np.zeros((n+1, 2))
    y[0] = y0
    for i in range(n):
        y[i+1] = M @ y[i]

    t = np.linspace(0, t_final, n+1)
    plt.plot(t, y[:,0], label=f"h={h}")

plt.legend()
plt.xlabel("t")
plt.ylabel("y1")
plt.title("Euler Method")
plt.grid()
plt.show()
```
