```python
import numpy as np
import time
import matplotlib.pyplot as plt

def mult(A, B):
    n = len(A)
    C = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k]*B[k][j]
    return C

sizes = [3]
for i in range(7):
    sizes.append(sizes[-1]*2)

t1 = []
t2 = []

for n in sizes:
    A = np.random.rand(n,n)
    B = np.random.rand(n,n)

    s = time.time()
    mult(A.tolist(), B.tolist())
    t1.append(time.time()-s)

    s = time.time()
    np.matmul(A,B)
    t2.append(time.time()-s)

x = range(1,9)

plt.plot(x, t1, marker='o', label='Naive')
plt.plot(x, t2, marker='o', label='NumPy')
plt.xlabel("Matrix Number")
plt.ylabel("Runtime")
plt.legend()
plt.show()
```

The Naive Runtime increased dramiatically as the martix number went beyond 6. Alternatively, the numpy runtime remained the same throughout. This graph represents the difference in efficiency for the two systems.
