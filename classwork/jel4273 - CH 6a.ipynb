```python
import numpy as np
import time

def gauss_seidel(A,b,x0,tol,max_iterations):
    n=len(b)
    x=x0.copy()
    start=time.time()
    for k in range(max_iterations):
        x_new=x.copy()
        for i in range(n):
            s1=sum(A[i][j]*x_new[j] for j in range(i))
            s2=sum(A[i][j]*x[j] for j in range(i+1,n))
            x_new[i]=(b[i]-s1-s2)/A[i][i]
        if np.linalg.norm(x_new-x,ord=np.inf)<tol:
            return x_new,k+1,time.time()-start
        x=x_new
    return x,max_iterations,time.time()-start

A=np.array([[4,-1,0,0],
            [-1,4,-1,0],
            [0,-1,4,-1],
            [0,0,-1,3]])

b=np.array([15,10,10,10])
x0=np.zeros_like(b)
tol=1e-6
max_iterations=100

solution,iterations,runtime=gauss_seidel(A,b,x0,tol,max_iterations)

print("Solution:",solution)
print("Iterations:",iterations)
print("Time:",runtime,"seconds")
```

```python
import numpy as np
import time

def jacobi(A,b,x0,tol,max_iterations):
    n=len(b)
    x=x0.copy()
    start=time.time()
    for k in range(max_iterations):
        x_new=x.copy()
        for i in range(n):
            s=sum(A[i][j]*x[j] for j in range(n) if j!=i)
            x_new[i]=(b[i]-s)/A[i][i]
        if np.linalg.norm(x_new-x,ord=np.inf)<tol:
            return x_new,k+1,time.time()-start
        x=x_new
    return x,max_iterations,time.time()-start

A=np.array([[4,-1,0,0],
            [-1,4,-1,0],
            [0,-1,4,-1],
            [0,0,-1,3]])

b=np.array([15,10,10,10])
x0=np.zeros_like(b)
tol=1e-6
max_iterations=100

solution,iterations,runtime=jacobi(A,b,x0,tol,max_iterations)

print("Solution:",solution)
print("Iterations:",iterations)
print("Time:",runtime,"seconds")
```

Both of the methods converged to the same solution, but the Gauss-Seidel method converged faster compated to the Jacobi method. Furthermore, the Gauss-Seidel method required only 3 iterations while the Jacobi method needed 4
