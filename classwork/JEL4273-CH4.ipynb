```python
import math
import numpy as np

def approximatepi(n):
    pi_sum = 0
    
    for i in range(n):
        term = (-1)**i / (2*i + 1)
        pi_sum += term
        
        if (i + 1) % 100 == 0:
            current_pi = 4 * pi_sum
            relative_error = abs(current_pi - np.pi) / np.pi
            print(f"Terms: {i+1}, Relative Error: {relative_error}")
    
    return 4 * pi_sum

print(approximatepi(5000))
```
