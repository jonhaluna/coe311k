```python
import numpy as np
import matplotlib.pyplot as plt

k = 1.0
x_start = 0
x_end = 10
y0 = 1.0
v0 = 0.0
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure()

for h in h_values:
    x = np.arange(x_start, x_end + h, h)
    y = np.zeros(len(x))
    v = np.zeros(len(x))

    y[0] = y0
    v[0] = v0

    for i in range(1, len(x)):
        y[i] = y[i-1] + h * v[i-1]
        v[i] = v[i-1] + h * (-k * y[i-1])

    plt.plot(x, y, label=str(h))

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 10
h = 0.01
x = np.arange(0, 10 + h, h)
y = np.zeros(len(x))
v = np.zeros(len(x))

y[0] = 1.0
v[0] = 0.0

for i in range(1, len(x)):
    y[i] = y[i-1] + h * v[i-1]
    v[i] = v[i-1] + h * (-k * y[i-1])

plt.plot(x, y)
plt.plot(x, v)
plt.show()
```

1. At around h=0.2 is where instability is initially seen.
2. Yes, it is becoming unstable since the amplitude is increasing instead of staying constant.
2. A larger k value makes the instablity occur earlier since the system oscillates faster.
4. Euler's method is not suitable for this problem if the h values increases. The h value must be small for Euler's method to be effective.
